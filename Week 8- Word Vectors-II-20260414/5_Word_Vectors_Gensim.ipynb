```python
from gensim.models import Word2Vec
model=Word2Vec(window=2,
         negative=20,
         min_count=5,vector_size=300)
```

```python
import pandas as pd
data=pd.read_csv('/content/train.csv')
```

```python
data.dropna(axis = 0, how ='any',inplace=True)
sentences=[row.split() for row in data['text']]
```

```python
model.build_vocab(sentences,progress_per=1000)
```

```python
model.train(sentences,epochs=30,total_examples=model.corpus_count)
```
